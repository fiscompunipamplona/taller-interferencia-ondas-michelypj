```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = 50. # cm
y1 = 50. # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")


for xx in range( len(x) ):
    for yy in range( len(y) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) ) 
        outfile.write("%0.1f " % (zeta0*sin(r1*k)))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("propagacion de la onda en x")
ylabel("propagacion de la onda en y")
imshow(data)
show()
```

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = float(input("Ingrese la posicion en x:")) # cm
y1 = float(input("Ingrese la posicion en y:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for yy in range( len(x) ):
    for xx in range( len(y) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r2 = sqrt( ((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) ) 
        zeta1=zeta0*sin(r1*k)
        zeta2=zeta0*sin(r2*k)
        outfile.write("%0.1f " % (zeta1))
        outfile.write("%0.1f " % (zeta2))
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower",extent=[0,100,0,100])
xlabel("propagacion de la onda en x(cm)")
ylabel("propagacion de la onda en y(cm)")
imshow(data)
show()
```

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = 50. # cm
y1 = 50. # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for xx in range( len(x) ):
    for yy in range( len(y) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r2 = sqrt( ((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) ) 
        zeta1=zeta0*sin(r1*k)
        zeta2=zeta0*sin(r2*k)
        outfile.write("%0.1f " % (zeta1+zeta2))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("propagacion de la onda en x(cm)")
ylabel("propagacion de la onda en y(cm)")
imshow(data)
show()
```
